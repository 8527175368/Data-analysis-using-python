```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('UFC.xlsx')
```

```python
data.head()
```

```python
data.info()
```

```python
data.sort_values('Fighter', inplace = True)
```

```python
data.head()
```

```python
data['Win Percentage'] = data['Win Percentage'] * 100
```

```python
data['Finish Percentage'] = data['Finish Percentage'] * 100
```

```python
data['Striking Accuracy'] = data['Striking Accuracy'] * 100
```

```python
data.rename(columns = {'Striking Accuracy' : 'Striking Accuracy Per'}, inplace = True)
```

```python
data.head(2)
```

```python
data['Fights'] = 0
```

```python
data.head(2)
```

```python
data['Fights'] = data['Wins'] + data['Losses'] + data['Draws']
```

```python
data.head()
```

```python
data['Fighting Style'].unique()
```

```python
data.loc[(data['Fighting Style'] == 'Unknown', 'Fighting Style')] = 'MMA'
```

```python
data.loc[(data['Fighting Style'] == 'Grappler', 'Fighting Style')] = 'Grappling'
```

```python
data.loc[(data['Fighting Style'] == 'Striker', 'Fighting Style')] = 'Striking'
```

```python
data.loc[(data['Fighter'] == 'Belal Muhammad', 'Fighting Style')] = 'Wrestling'
```

```python
data.head(10)
```

```python
third_column = data.pop('Fights')
```

```python
third_column
```

```python
data.insert(2, 'Fights', third_column)
```

```python
data.head()
```

# Q-1 : Which division has the most knockouts?

```python
ko_by_division = data.groupby('Weight Class').sum()['KO TKO'].reset_index()
```

```python
ko_by_division
```

```python
colors = np.array([0, 10, 20, 30, 40, 50, 60, 70])
```

```python
plt.figure(figsize = (15, 10))
plt.bar(ko_by_division['Weight Class'], ko_by_division['KO TKO'], color = '#FF1E1E', width =0.7)
plt.title('Knockouts by division')
plt.xlabel('Weight Class')
plt.ylabel('KO TKO')
plt.show()
```

# Q-2 : Which division has the most submissions ?

```python
sub_by_division = data.groupby('Weight Class').sum()['Submissions'].reset_index()
```

```python
sub_by_division
```

```python
plt.figure(figsize = (12, 9))
plt.barh(sub_by_division['Weight Class'], sub_by_division['Submissions'], color = '#187DFF')
plt.xlabel('Submissions')
plt.ylabel('Weight Class')
plt.title('Submissions by division')
plt.show()
```
